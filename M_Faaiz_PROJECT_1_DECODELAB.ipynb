```python
import pandas as pd
import numpy as np


file_path = "Dataset for Data Analytics.xlsx"


df = pd.read_excel(file_path, sheet_name='Sheet1')


print("--- Initial Data Summary ---")
print(f"Total Rows: {df.shape[0]}, Total Columns: {df.shape[1]}")
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
print("--- Checking for Missing Values ---")
print(df.isnull().sum())

df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

print("\n--- After Handling Missing Values ---")
print(f"Missing values in CouponCode now: {df['CouponCode'].isnull().sum()}")
```

```python
print("--- Checking for Duplicates ---")
total_duplicates = df.duplicated().sum()
order_id_duplicates = df.duplicated(subset=['OrderID']).sum()

print(f"Total fully duplicate rows: {total_duplicates}")
print(f"Duplicate OrderIDs found: {order_id_duplicates}")


df = df.drop_duplicates(subset=['OrderID'], keep='first')
print("\nDuplicates handled. Total rows remaining:", len(df))
```

```python
print("--- Formatting and Validation ---")

df['Date'] = pd.to_datetime(df['Date'])

text_columns = ['OrderID', 'CustomerID', 'Product', 'ShippingAddress', 'PaymentMethod', 'OrderStatus', 'ReferralSource']
for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

df['CalculatedPrice'] = df['Quantity'] * df['UnitPrice']
price_mismatch = df[(df['TotalPrice'] - df['CalculatedPrice']).abs() > 0.01]

print(f"Number of rows with price mismatches: {len(price_mismatch)}")

df = df.drop(columns=['CalculatedPrice'])
```

```python
output_file = "Cleaned_Dataset_Project_1.xlsx"
df.to_excel(output_file, index=False)

print(f"🎉 Success! Your clean dataset has been saved as: {output_file}")
print("You can download this file from Google Colab's file explorer side panel.")
```
